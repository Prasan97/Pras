```python
import numpy as np
```

```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
Pras_data=pd.read_csv('Mall_Customers.csv')
```

```python
Pras_data
```

```python
X=Pras_data.iloc[:,[3,4]].values
```

```python
X
```

```python
from sklearn.cluster import KMeans
wcss=[]

```

```python
for i in range (1,21):
    k_means=KMeans(i,init='k-means++',random_state=100)
    k_means.fit(X)
    wcss.append(k_means.inertia_)
```

```python
plt.plot(range(1,21),wcss)
```

```python
k_means=KMeans(5,init='k-means++',random_state=100)
```

```python
result=k_means.fit_predict(X)
```

```python
plt.scatter(X[result==0,0], X[result==0,1], c='red')
plt.scatter(X[result==1,0], X[result==1,1], c='pink')
plt.scatter(X[result==2,0], X[result==2,1], c='green')
plt.scatter(X[result==3,0], X[result==3,1], c='yellow')
plt.scatter(X[result==4,0], X[result==4,1], c='cyan')
plt.scatter(k_means.cluster_centers_[:,0],k_means.cluster_centers_[:,1],s=100,c='black')
```
